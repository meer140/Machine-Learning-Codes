```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
x, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x[:,0],x[:,1],s=30,c=y)
```

```python
def perceptron(x,y):
    x = np.insert(x,0,1,axis=1)
    w = np.ones(x.shape[1])
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = sigmoid(np.dot(x[j],w))
        w = w + lr*(y[j] - y_hat)*x[j]

    return w[0],w[1:]
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
intercept,coef = perceptron(x,y)
```

```python
intercept,coef
```

```python
m = -coef[0]/coef[1]
b = - intercept/coef[1]
```

```python
x_input = np.linspace(-3,3,100)
y_output = m * x_input + b
```

```python
plt.figure(figsize=(8,4))
plt.plot(x_input,y_output,color='red')
plt.scatter(x[:,0],x[:,1],c=y,s=20)
plt.ylim(-3,3)
```

```python
lor = LogisticRegression()
lor.fit(x,y)
```

```python
m2 = -(lor.coef_[0][0]/lor.coef_[0][1])
b2 = -(lor.intercept_/lor.coef_[0][1])
m2,b2
```

```python
x_input2 = np.linspace(-3,3,100)
y_input2 = m2 * x_input2 + b2
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_output,color='red',linewidth=2)
plt.plot(x_input2,y_input2,color='black',linewidth=2)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=50)
plt.ylim(-3,3)
```
