```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import cross_val_score
```

```python
x,y = make_moons(n_samples=500, noise=0.15,random_state=1)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python
clf.fit(x,y)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x, y, clf, legend=2)
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=10))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
def plot_graph(x,y,deg=1):

    model = Pipeline([
        ("poly", PolynomialFeatures(degree=deg)),
        ("lr", LogisticRegression())
    ])
    
    # Fit pipeline
    model.fit(x, y)
    
    # Accuracy with cross-validation
    accuracy = np.mean(cross_val_score(model, x, y, scoring='accuracy', cv=10))
    print("Accuracy:", accuracy)
    
    # Plot decision regions (pass original x, pipeline handles transformation)
    plot_decision_regions(x, y, clf=model, legend=2)
    plt.show()
```

```python
plot_graph(x,y)
```

```python
plot_graph(x,y,2)
```

```python
plot_graph(x,y,3)
```
